```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path = '/home/vbertalan/Downloads/Projetos/Nezha/rca_data - fault suffering data/OnlineBoutique-2022-08-22/trace/03_53_trace.csv'  # Replace with your file path
df = pd.read_csv(file_path)

# Initialize a directed graph
dag = nx.DiGraph()

# Add edges to the graph from the DataFrame
for index, row in df.iterrows():
    span_id = row['SpanID']
    parent_id = row['ParentID']
    
    # Add the edge only if ParentID is not null
    if pd.notnull(parent_id):
        dag.add_edge(parent_id, span_id)

# Print the nodes and edges in the DAG
print("Nodes in the DAG:", dag.nodes())
print("Edges in the DAG:", dag.edges())

# Check if the graph is a DAG
if nx.is_directed_acyclic_graph(dag):
    print("The graph is a DAG.")
else:
    print("The graph is not a DAG.")

# Visualize the DAG
plt.figure(figsize=(10, 8))  # Set the figure size
pos = nx.spring_layout(dag)  # Use a layout for better visualization
nx.draw(dag, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=10, font_weight='bold', arrows=True)

# Show the plot
plt.title("Directed Acyclic Graph (DAG)")
plt.show()
```
